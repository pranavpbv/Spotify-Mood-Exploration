```python
#pranav
collection.find_one()
```

```python
#pranav
collection.count_documents({})
```

```python
#pranav
pipeline = [
    {"$group": {"_id":"$genre", "count": {"$sum": 1}}},
    {"$sort": {"count": -1}},
    {"$project": {"_id": 0, "genre": "$_id", "count": 1}}
]

results=collection.aggregate(pipeline)
print("Top 7 genres")
for i, result in enumerate(results, 1):
  print(f"{i}. {result["genre"]}: {result["count"]}")
```

```python
# Carson

top_loud = collection.find(
    {},
    {"_id": 0, "artist_name": 1, "track_name": 1, "loudness": 1}
).sort("loudness", pymongo.DESCENDING).limit(10)

print("Top 10 Loudest Songs")
for i, song in enumerate(top_loud, 1):
  print(f"{i}. \"{song["track_name"]}\" by {song["artist_name"]}: {round(song["loudness"], 3)} loudness")
```

```python
# Carson
pipeline = [
    {"$group": {
        "_id": "$artist_name",
        "avg_length": {"$avg": "$len"}
    }},
    {"$sort": {"avg_length": -1}},
    {"$limit": 10},
    {"$project": {
        "_id" : 0,
        "artist_name": "$_id",
        "avg_length" : 1
    }}
]

results = collection.aggregate(pipeline)
print("Top 10 Artists With Longest Lyrics")
for result in results:
  print(f"{result["artist_name"]}: {result["avg_length"]}")
```

```python
# Pranav
pipeline = [{
    "$addFields": {
        "bedtime_score": {
            "$subtract": [
                {
                    "$add": [
                        {"$multiply": ["$night/time", 3]},
                        {"$multiply": ["$family/gospel",3]},
                        {"$multiply": ["$family/spiritual", 3]},
                        "$acousticness"
                    ]
                },
                {
                  "$add": [
                      {"$multiply": ["$obscene", 4]},
                      {"$multiply": ["$violence",3]},
                      {"$multiply": ["$energy", 1.5]},
                  ]
                }
            ]
        }
    }

},{
    "$sort": {"bedtime_score": -1}
            },
            {
                "$limit": 10
            }
            ,
{

    "$project": {
        "_id": 0,
        "artist_name": 1,
        "track_name": 1,
        "bedtime_score": 1
}
}
]

results = collection.aggregate(pipeline)
print("Top 10 Bedtime Songs")
for result in results:
  print(f"{result['track_name']} by {result['artist_name']}: {round(result['bedtime_score'],3)}")



```
